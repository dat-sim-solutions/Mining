## Bishop method for the safety factor $F_s$. Tailing Dam
The Tailing Dam has a cross section 2D. Choose the tallest/steepest part of the dam. \
The "Slices" (1D Columns): You divide that 2D cut into 20–30 vertical columns (slices)

Stress Transformation Equations, it is a 2D aproximation, in the XY plane (horizontal-vertical surface that is perpendicular to the slope surface). If you have the major principal stress ($\sigma_1$ pointing to Y direction) and minor principal stress ($\sigma_3$ pointing to X direction), the stresses on any plane at an angle $\theta$ (in the soil, choose a point on that plane and you have a normal and a shear stress) are:
$$\sigma_\theta = \frac{\sigma_1 + \sigma_3}{2} + \frac{\sigma_1 - \sigma_3}{2} \cos(2\theta)$$
$$\tau_\theta = \frac{\sigma_1 - \sigma_3}{2} \sin(2\theta)$$
these are parametric equations for a circle: \
Center: Located at $(\frac{\sigma_1 + \sigma_3}{2}, 0)$. \
Radius ($R$): Equal to the maximum shear stress, $\frac{\sigma_1 - \sigma_3}{2}$. \
Parametric equations of circle: $x=h+r\cos(\theta)$ and $y=k+r\sin(\theta)$.

1) Mohr-Coulomb equation: $s$ (= $\tau_\theta$ in the previous symbols) $$s = c' + \sigma' \tan \phi'$$\
$s$ (Shear Strength): The maximum resistance a soil or rock can offer before it slides or collapses.\
$c'$ (Effective Cohesion): The "glue." It is the strength of the material when there is zero pressure on it\
$\sigma'$ (Effective Normal Stress): The "clamping force." It is the weight of the material above the failure plane, minus the pore water pressure. \
$\tan \phi'$ (Coefficient of Internal Friction): Like a "friction coefficient" for soil. $\phi'$ is the angle of internal friction—essentially how "rough" the soil particles are.

2) Factor of safety: $  F_s= \dfrac{s}{\tau_i}$ = stress available / stress being used \
 $  \tau_i = \dfrac{s}{F_s}$  \
 stress = force / area . Along the surface of the slice and the slope.\
 $ \tau_i = \dfrac{T_i}{l_i \times 1}= \dfrac{T_i}{l_i}$; where $l_i$ is the lentgh of the base (along the slope), the width is 1 m (into the screen). \
 $  \sigma'$ in the form of forces is $ \dfrac{N_i}{l_i \times 1} - u_i = \dfrac{N_i}{l_i} - u_i $. We let $c'$ in the form of stress.

$$ \frac{T_i}{l_i} = \frac{1}{F_s} [c' + (\frac{N_i}{l_i}  - u_i) \tan \phi']$$

The shear force $T_i$ (Mobilized Stress, the actual force being used to keep the slope from moving; is not the available force) is defined by the Shear Strength (max resistance) divided by the Factor of Safety ($F_s$):
$$T_i = \frac{1}{F_s} [c' \cdot l_i + (N_i - u_i \cdot l_i) \tan \phi']$$

3) $\sum F_y = 0$, $$N_i \cos \alpha_i + T_i \sin \alpha_i = W_i$$
We have 2 equations with 2 variables: $T_i$ and $N_i$, so we can solve for $T_i$ and $N_i$.\
a) Substitute the $T_i$ expression into the vertical equilibrium equation and solve for $N_i$:
$$N_i = \frac{W_i - \frac{1}{F_s} (c' \cdot l_i - u_i \cdot l_i \tan \phi') \sin \alpha_i}{\cos \alpha_i + \frac{\sin \alpha_i \tan \phi'}{F_s}}$$
b) $$T_i = \frac{c' l_i + (N_i - u_i l_i) \tan \phi'}{F_s}$$
c) $$\sigma_n = \frac{N_i}{l_i}$$ Once your app calculates the final $F_s$, you can use this formula to map the Normal Stress Distribution along the entire failure surface of the dam. If $\sigma_n$ becomes negative, the dam is experiencing tension

4) Summing moments about the center of the circular slip surface ($O$) for the entire mass: 
$$\sum W_i \sin \alpha_i \cdot R = \sum T_i \cdot R$$
Substitute the expression for $T_i$ and $N_i$. After significant algebraic manipulation and we reach the final Bishop Equation:

$$F_s = \frac{\sum \left[ c' b_i + (W_i - u_i b_i) \tan \phi' \right] \cdot \frac{1}{m_{\alpha, i}}}{\sum W_i \sin \alpha_i}$$

$$m_{\alpha, i} = \cos \alpha_i + \frac{\sin \alpha_i \tan \phi'}{F_s}$$
$ c':$ effective cohesion, from the Mohr-Coulomb equation ($s = c' + \sigma' \tan \phi'$). For clean, dry sand, $c'$ is usually 0\
$ b_i:$ Width of the slice. At the base there is a triangle where $l_i \cos \alpha_i = b_i$ \
$W_i$: Total weight of the slice \
$ u_i:$ the pressure of the water trapped between the soil particles (pores) in slice $i$, it pushes the soil grains apart (liquefaction). It "buoys" the soil and reduces friction \
$\phi':$ the angle of internal friction. It represents the resistance of soil grains to sliding over each other. The steeper the friction angle, the more stable the dam\
$ \alpha:$ The angle of the slip surface at the bottom of that specific slice.\
The prime symbol ($'$) signifies that this value is calculated using Effective Stress (Total Stress minus Pore Water Pressure)

5) Normal stress, given the current Factor of Safety. $$\sigma_{n,i} = \frac{N_i}{l_i} = \frac{W_i - \frac{1}{F_s}(c' l_i - u_i l_i \tan \phi') \sin \alpha_i}{l_i (\cos \alpha_i + \frac{\sin \alpha_i \tan \phi'}{F_s})}$$
Mobilized shear stress ($\tau_m$). This is the shear stress actually acting on the base of the slice to prevent it from sliding. By definition of the Factor of Safety:$$\tau_{m,i} = \frac{\text{Shear Strength}}{\text{Safety Factor}} = \frac{c' + (\sigma_{n,i} - u_i) \tan \phi'}{F_s}$$
